Q1) What is a parameter?
- A parameter is a variable that defines a model’s internal configuration and is learned from the data during training. For example, in linear regression, the slope and intercept are parameters. These values determine how the model makes predictions.

Q2)  What is correlation?
 What does negative correlation mean?
 - Correlation measures the statistical relationship between two variables. It tells us whether one variable changes in relation to another. Correlation values range from -1 to +1:

- +1 means perfect positive relationship.
- -1 means perfect negative relationship.

- Negative correlation means that when one variable increases, the other decreases. For example, the number of hours spent watching TV might negatively correlate with academic performance.

Q3) Define Machine Learning. What are the main components in Machine Learning?
- Machine Learning (ML) is a field of artificial intelligence where computers learn patterns from data without being explicitly programmed.
The main components are:

- Dataset (input data for learning).
- Features (independent variables used for prediction)
- Model/Algorithm (mathematical representation for learning patterns)
- Loss function (measures how far predictions are from actual results)
- Optimizer (method to update parameters)
- Training & Testing process (to evaluate performance).

Q4)  How does loss value help in determining whether the model is good or not?
- The loss value represents the error between predicted and actual results. A lower loss means the model is performing well. If the loss is high, it means the model is making poor predictions. Monitoring loss during training helps us know if the model is improving.

Q5)  What are continuous and categorical variables?
- Continuous variables are numerical and can take infinite values within a range (e.g., height, weight, temperature).
- Categorical variables represent discrete groups or categories (e.g., gender, colors, country names).

Q6) How do we handle categorical variables in Machine Learning? What are the common techniques?
- Categorical variables cannot be directly fed into models, so we encode them into numbers. Common techniques are:
- Label Encoding (assigning each category a unique integer)
- One-Hot Encoding (creating binary columns for each category)
- Ordinal Encoding (for ordered categories, e.g., small < medium < large).

Q15)  What is causation? Explain difference between correlation and causation with an example.

- Causation means one variable directly affects another. Correlation only shows a relationship, but not necessarily cause-effect.
Example: Ice cream sales and drowning cases are correlated (both increase in summer). But ice cream does not cause drowning. The cause is high temperature.

Q16) What is an Optimizer? What are different types of optimizers? Explain each with an example.
- An optimizer is an algorithm used to adjust model parameters to minimize the loss function.
Common optimizers:

- Gradient Descent: updates parameters step by step in the direction of - steepest decrease of loss.
- Stochastic Gradient Descent (SGD): uses one random sample at a time to update parameters, faster but noisier.
- Adam: combines momentum and adaptive learning rate, widely used in deep learning.
- RMSProp: scales learning rate based on past gradients.

Q21)  What is feature scaling? How does it help in Machine Learning?
- Feature scaling transforms numerical values into a standard range (e.g., 0–1 or mean=0, std=1). It prevents features with large values from dominating models. Example: Age (20–60) vs Salary (20,000–100,000). Scaling ensures fair comparison.

- Data encoding is the process of converting categorical variables into numerical values so ML models can understand them.

One-Hot Encoding -> creates binary columns
